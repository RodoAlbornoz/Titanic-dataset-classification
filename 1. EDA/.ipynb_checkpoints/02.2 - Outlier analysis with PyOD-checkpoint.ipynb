# Using PyOD

Import the models that we are going to use

```python
from pyod.models.knn import KNN
from pyod.models.lof import LOF
from pyod.models.iforest import IForest
from pyod.models.hbos import HBOS
from pyod.models.abod import ABOD
from pyod.models.cblof import CBLOF
from pyod.models.feature_bagging import FeatureBagging
from pyod.models.mcd import MCD
from pyod.models.ocsvm import OCSVM
from pyod.models.pca import PCA
from pyod.models.lscp import LSCP
from pyod.models.sos import SOS
from pyod.models.cof import COF
from pyod.models.sod import SOD
```

Define variables for our models. We found in our other outlier analysis that about 30% of our dataset are outliers, so we are going to use that number for our models.

```python
outliers_fraction = 0.30
random_state = 42
detector_list = [LOF(n_neighbors = 5), LOF(n_neighbors = 10), LOF(n_neighbors = 15),
                 LOF(n_neighbors = 20), LOF(n_neighbors = 25), LOF(n_neighbors = 30),
                 LOF(n_neighbors = 35), LOF(n_neighbors = 40), LOF(n_neighbors = 45),
                 LOF(n_neighbors = 50)]
feature_model_names = []
```

Define our 16 models

```python
models = {
    'K-Nearest Neighbors (KNN)': KNN(contamination = outliers_fraction),
    'Median KNN' : KNN(method = 'median', contamination = outliers_fraction),
    'Average KNN': KNN(method = 'mean', contamination = outliers_fraction),
    'Local Outlier Factor (LOF)': LOF(n_neighbors = 35, contamination = outliers_fraction),
    'Isolation Forest': IForest(contamination = outliers_fraction, random_state = random_state),
    'Histogram-Based Outlier Score (HBOS)': HBOS(contamination = outliers_fraction),
    'Angle-Based Outlier Detection (ABOD)': ABOD(contamination = outliers_fraction),
    'Cluster-Based Local Outlier Factor (CBLOF)': CBLOF(contamination = outliers_fraction, check_estimator = False, random_state = random_state),
    'Feature Bagging': FeatureBagging(LOF(n_neighbors = 35), contamination = outliers_fraction, random_state = random_state),
    'Minimum Covariance Determinant (MCD)': MCD(contamination = outliers_fraction, random_state = random_state),
    'One-Class SVM (OCSVM)': OCSVM(contamination = outliers_fraction),
    'Principal Component Analysis (PCA)': PCA(contamination = outliers_fraction, random_state = random_state),
    'Stochastic Outlier Selection (SOS)': SOS(contamination = outliers_fraction),
    'Locally Selective Combination (LSCP)': LSCP(detector_list, contamination = outliers_fraction, random_state = random_state),
    'Connectivity-Based Outlier Factor (COF)': COF(n_neighbors = 35, contamination = outliers_fraction),
    'Subspace Outlier Detection (SOD)': SOD(contamination = outliers_fraction),
}
```

Work only with the features we're interested in, and fill null values with zeroes just for the purpose of outlier analysis, since some models don't accept null values.

```python
ds_aux = ds_work.copy()
ds_aux.fillna(0, axis = 1, inplace = True)
X1 = ds_aux[["Age", "Fare"]].values
X2 = ds_aux[["Age", "SibSp"]].values
X3 = ds_aux[["Age", "Parch"]].values
X4 = ds_aux[["Fare", "SibSp"]].values
X5 = ds_aux[["Fare", "Parch"]].values
X6 = ds_aux[["SibSp", "Parch"]].values
```

Define the dimensions of our graphs

```python
num_models = len(models)
num_rows = 4
num_cols = (num_models + 1) // num_rows
```

Define our function to create each model, plot outliers for each model, and add flags in our dataset to know whether certain row is considered an outlier (1) or not (0) for each model

```python
def plot_outliers_for_features(X, dataset, feat1, feat2, feature_model_names):
    fig, axs = plt.subplots(num_rows, num_cols, figsize = (5 * num_rows,5 * num_cols))
    for i, (model_name, model) in enumerate(models.items()):
        model.fit(X)
        y_pred = model.predict(X)

        name_feature_by_model = 'Is_outlier_for_' + model_name
        feature_model_names.append(name_feature_by_model)
        dataset[name_feature_by_model] = y_pred.tolist()

        inliers = X[y_pred == 0]
        outliers = X[y_pred == 1]

        row_idx = i // num_cols
        col_idx = i % num_cols
        axs[row_idx, col_idx].scatter(inliers[:, 0], inliers[:, 1], c = 'blue', label = 'Inliers', edgecolor = 'k')
        axs[row_idx, col_idx].scatter(outliers[:, 0], outliers[:, 1], c = 'red', label = 'Outliers', edgecolor = 'k')
        axs[row_idx, col_idx].set_title(f'Outliers {model_name}')
        axs[row_idx, col_idx].set_xlabel(feat1)
        axs[row_idx, col_idx].set_ylabel(feat2)
        axs[row_idx, col_idx].legend()
        
    feature_model_names.clear()
        
    plt.tight_layout()
    plt.show()
```

We're going to plot each variable against the others

# Age vs Fare

```python
ds_aux2 = ds_aux.copy()
plot_outliers_for_features(X1, ds_aux2, "Age", "Fare", feature_model_names)
```

# Age vs SibSp

```python
ds_aux3 = ds_aux.copy()
plot_outliers_for_features(X2, ds_aux3, "Age", "SibSp", feature_model_names)
```
