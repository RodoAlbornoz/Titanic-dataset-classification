## Frequency values

Let's see the frequency values for each feature, except for:
* Age and Fare, since they are continuous
* PassengerId, since it takes unique values
* Name, since it contains all unique string values

```python
ds_aux = ds_work.copy()
ds_aux.drop(['PassengerId', 'Name', 'Age', 'Fare'], axis = 1, inplace = True)

for col in ds_aux.columns:
  print('Variable:', col)
  print(f"{ds_aux[col].value_counts(dropna = False).to_string()}")
  print('_________________________') 
```

This tells us:
* People tended to die rather than survive
* Males tended to die more than females

## Graph features

Now that we have explored the possible values, range and other things for each feature, let's plot each one with histograms, excluding PassengerId, Name, Ticket and Cabin, since they have a lot of unique values

```python
ds_aux2 = ds_work.copy()
ds_aux2.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1, inplace = True)

num_columns = len(ds_aux2.columns)
columns_per_row = 2
num_rows = (num_columns + columns_per_row - 1) // columns_per_row

fig, axes = plt.subplots(num_rows, columns_per_row, figsize = (15, 5 * num_rows))
axes = axes.flatten()

for i, column in enumerate(ds_aux2.columns):
    ax = axes[i] 
    sns.histplot(ds_aux2[column], ax = ax, bins = 70)
    ax.set_title(column) 

plt.tight_layout()
plt.show()
```

Here, we can see some interesting things:
* Most common ages were around 20 and 40 years
* Most common Fare values were below 100

We can also see some possible outliers:
* There isn't much people that is more than 60 years old
* SibSp values above 1 seem to be much less common
* Parch values above 2 seem to be much less common
* Some Fare values seem to be very isolated. For example, there are some fares around 200 and 300. There is also a fare a little bit greater than 500.

We are going to analyze and treat these outliers later.

## Repeated Ticket Values

Now, let's check values for Ticket. There are 681 unique values for it, so let's check repeated values.

```python
ds_work[ds_work.duplicated(['Ticket'], keep = False)].sort_values(by = "Ticket")
```

Given some last names are repeated, we could conclude that some repeated tickets come from members of certain families.

It could also be that all people who shared a particular ticket either died or survived. Let's make a scatterplot to check if there is a visible relationship and it is true.

```python
plt.figure(figsize = (5, 23))
sns.scatterplot(data = ds_work[ds_work.duplicated(['Ticket'], keep = False)].sort_values(by = "Ticket"), x = "Survived" , y = "Ticket")
plt.show()
```

This tells us that most people who shared a particular ticket died and some others survived. Therefore, it doesn't seem to be a relevant correlation, and since Ticket doesn't give us relevant information about whether a passenger died or not, we can eliminate it from our dataset.

## Repeated Cabin values

Now, let's look at all repeated values for Cabin (Excluding null values)

```python
ds_repeated_cabin = ds_work[ds_work.duplicated(['Cabin'], keep = False)].sort_values(by = "Ticket").dropna()
ds_repeated_cabin
```

It seems like most people who shared Cabin, also shared Ticket. Let's check if that is true. 

```python
ds_repeated_cabin_and_ticket = ds_work[ds_work.duplicated(['Cabin'], keep = False)][ds_work.duplicated(['Ticket'], keep = False)].sort_values(by = "Ticket").dropna()
ds_repeated_cabin_and_ticket
```

Let's check how many people share Cabin

```python
ds_repeated_cabin.dropna().shape[0]
```

Let's also check how many people share both Ticket and Cabin

```python
ds_repeated_cabin_and_ticket.dropna().shape[0]
```

So, out of 95 people who shared Cabin, 76 (80%) also shared Ticket. But since Cabin has a 77% of null values which don't give us useful information, and Ticket is going to be deleted for not having useful information, we could also eliminate it from our dataset too.

# Check duplicated values

Let's check if there are duplicated values in our dataset. Since PassengerId and Name take all unique values, let's exclude them

```python
ds_aux3 = ds_work.copy()
ds_aux3[ds_aux3.duplicated(subset = ds_aux3.drop(["PassengerId", "Name"], axis = 1, inplace = True))]
```

We find some repeated rows. But since they don't share name, and its Age and Cabin values contain null values, we cannot conclude that a passenger is repeated. For example: 

```python
ds_work[ds_work["PassengerId"].isin([202, 325])]
```

Even though they share values for almost all features, they have different names. Maybe they are father/son or siblings.

Therefore, there are no duplicated values, and we can distinguish these rows thanks to their name.

We explored our dataset. We checked for feature values, outliers, correlated and irrelevant features.
